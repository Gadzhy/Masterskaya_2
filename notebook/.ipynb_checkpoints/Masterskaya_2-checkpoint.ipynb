# Мастерская_2

В данном проекте я построю модель предсказания вероятности совершения покупки клиентами интернет-магазина в течение ближайших 90 дней.  

В колонках датасета содержатся следующие данные:  
    
  **apparel-purchases**

- `client_id` — идентификатор клиента
- `quantity` — количество единиц товара
- `price` — цена товара
- `category_ids` — идентификаторы категорий
- `date` — дата покупки
- `message_id` — идентификатор сообщения из рассылки  

  **apparel-messages**

- `bulk_campaign_id` — идентификатор рассылки
- `client_id` — идентификатор клиента
- `message_id` — идентификатор сообщения
- `event ` — действие с сообщением (отправлено, открыто, покупка…)
- `channel` — канал рассылки
- `date` — дата действия
- `created_at` — дата-время полностью

  **target**
  
- `client_id` — идентификатор клиента
- `target ` — клиент совершил покупку в целевом периоде

  **full_campaign_daily_event**
  
- `date` — дата
- `bulk_campaign_id` — идентификатор рассылки
- `count_event*` — общее количество каждого события event
- `nunique_event* ` — количество уникальных client_id в каждом событии

  **full_campaign_daily_event_channel**
  
- `date` — дата
- `bulk_campaign_id` — идентификатор рассылки
- `count_event*_channel* ` — общее количество каждого события по каналам
- `nunique_event*_channel*`- количество уникальных client_id по событиям и каналам

# Подготовка данных

```python
purchases = pd.read_csv('../data/apparel-purchases.csv', parse_dates=['date'])
```

```python
messages = pd.read_csv('../data/apparel-messages.csv', parse_dates=['date', 'created_at'])
```

```python
full_campaign_channel = pd.read_csv('../data/full_campaign_daily_event_channel.csv', parse_dates=['date'])
```

```python
df_name = ['purchases', 'messages', 'target', 'full_campaign', 'full_campaign_channel']
```

```python
# Выведем первичную информацию о датасетах
for df, name in zip(df, df_name):
    print(f"Данные из датафрейма '{name}':")
    display(df.head())
    print(df.info())
```

```python
# Сначала создаем purchases_exploded для подсчета уникальных категорий
purchases_exploded = purchases.copy()
```

```python
# Преобразуем category_ids в список, если это строка
if isinstance(purchases['category_ids'].iloc[0], str):
    purchases_exploded['category_ids'] = purchases_exploded['category_ids'].apply(
        lambda x: x.strip('[]').replace("'", "").replace('"', '').split(',')
    )
```

```python
# Разделяем категории и считаем уникальные
unique_cats = purchases_exploded.explode('category_ids').groupby('client_id')['category_ids'].nunique().reset_index()
unique_cats.columns = ['client_id', 'unique_categories']
```

```python
# Агрегация purchases по client_id
purchases_agg = purchases.groupby('client_id').agg({
    'price': ['sum', 'mean', 'count'],  # общая сумма, средний чек, количество покупок
    'date': 'max',  # последняя покупка
    'quantity': 'sum'  # общее количество товаров
}).reset_index()
```

```python
# Переименование колонок
purchases_agg.columns = [
    'client_id', 
    'total_spent', 
    'mean_spent', 
    'purchase_count', 
    'last_purchase_date', 
    'total_quantity'
]
```

```python
purchases_agg.info()
```

```python
purchases_agg.duplicated().sum()
```

```python
# Объединяем с уникальными категориями
purchases_agg = purchases_agg.merge(unique_cats, on='client_id', how='left')
purchases_agg['unique_categories'] = purchases_agg['unique_categories'].fillna(0)
```

```python
# Фильтруем только события open и click
messages_filtered = messages[messages['event'].isin(['open', 'click'])]
```

```python
# Агрегация messages по client_id
messages_agg = messages_filtered.groupby('client_id').agg({
    'event': lambda x: (x == 'open').sum(),  # количество открытий
    'date': ['max', 'min'],  # последнее и первое сообщение
    'channel': 'nunique',  # количество уникальных каналов
    'message_id': 'count'  # общее количество сообщений
}).reset_index()
```

```python
# Переименование колонок
messages_agg.columns = [
    'client_id', 
    'open_count', 
    'last_message_date', 
    'first_message_date', 
    'unique_channels', 
    'total_messages'
]
```

```python
messages_agg.info()
```

```python
messages_agg.duplicated().sum()
```

```python
# Добавляем количество кликов
click_count = messages_filtered[messages_filtered['event'] == 'click'].groupby('client_id').size().reset_index(name='click_count')
messages_agg = messages_agg.merge(click_count, on='client_id', how='left')
messages_agg['click_count'] = messages_agg['click_count'].fillna(0)
```

```python
final_df = target.copy()
```

```python
# Присоединяем purchases features
final_df = final_df.merge(purchases_agg, on='client_id', how='left')
```

```python
# Присоединяем messages features
final_df = final_df.merge(messages_agg, on='client_id', how='left')
```

```python
# Создаем бинарный признак has_messages
final_df['has_messages'] = final_df['open_count'].notna().astype('int')
print("Распределение клиентов по признаку has_messages:")
print(final_df['has_messages'].value_counts())
```

```python
final_df = final_df.fillna({
    'last_purchase_date': pd.NaT,
    'last_message_date': pd.NaT,
    'first_message_date': pd.NaT 
})
```

```python
# Разница между последним сообщением и последней покупкой
final_df['days_since_last_purchase'] = (final_df['last_message_date'] - final_df['last_purchase_date']).dt.days
```

```python
# Длительность взаимодействия с клиентом
final_df['interaction_duration_days'] = (final_df['last_message_date'] - final_df['first_message_date']).dt.days
```

```python
# Средняя частота сообщений
final_df['avg_messages_per_day'] = final_df['total_messages'] / final_df['interaction_duration_days'].replace(0, 1)
```

```python
# Коэффициент открытий
final_df['open_rate'] = final_df['open_count'] / final_df['total_messages'].replace(0, 1)
```

```python
# Заполняем пропуски в счетчиках
count_columns_to_fill = ['open_count', 'click_count', 'total_messages', 'unique_channels']
final_df[count_columns_to_fill] = final_df[count_columns_to_fill].fillna(0)

# Зполним пропуски в признаках основанных на датах, если такие имеются
final_df['days_since_last_purchase'] = final_df['days_since_last_purchase'].fillna(9999)
final_df['interaction_duration_days'] = final_df['interaction_duration_days'].fillna(0)
final_df['avg_messages_per_day'] = final_df['avg_messages_per_day'].fillna(0)
final_df['open_rate'] = final_df['open_rate'].fillna(0)
```

# Обучение и анализ моделей

```python
# Выделим целевую переменную
X = final_df.drop(columns=['target','client_id'], axis=1)
y = final_df['target']
```

```python
# Разделим датасет на выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = RANDOM_STATE)
```

```python
# Определим категориальные и числовые признаки
cat_features = [] 
num_features = ['total_spent', 'mean_spent', 'total_quantity', 'unique_categories', 'open_count',\
                'click_count', 'interaction_duration_days',\
                'avg_messages_per_day', 'open_rate', 'has_messages']
```

```python
# Пайплайн для категориальных признаков
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent', fill_value='unknown')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))
])
```

```python
# Пайплайн для числовых признаков
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value=0)), 
    ('scaler', StandardScaler())
])
```

```python
# Объединяем трансформеры в ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_features),
        ('cat', cat_transformer, cat_features)
    ])
```

```python
# Создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced'))
])
```

```python
param_grid = [
    # Logistic Regression
    {
        'model': [LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')],
        'model__C': [0.001, 0.01, 0.1, 1, 10, 100],
        'model__solver': ['lbfgs', 'newton-cg', 'sag']
    },
    
    # Random Forest
    {
        'model': [RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1, class_weight='balanced')],
        'model__n_estimators': [50, 100, 200],
        'model__max_depth': [10, 20, 30, None],
        'model__max_features': ['sqrt', 'log2', 0.5, 0.8],
        'model__min_samples_split': [2, 5, 10]
    },
    
    # CatBoost
    {
        'model': [CatBoostClassifier(random_state=RANDOM_STATE, verbose=0)],
        'model__iterations': [100, 200, 300, 500],
        'model__depth': [4, 5, 6, 7, 8],
        'model__learning_rate': [0.01, 0.03, 0.05, 0.1],
        'model__l2_leaf_reg': [1, 3, 5, 7, 10],
        'model__grow_policy': ['SymmetricTree', 'Depthwise'],
        'model__min_data_in_leaf': [1, 5, 10, 20],
        'model__random_strength': [1, 2, 3],
        'model__bagging_temperature': [0, 0.5, 1],
        'model__border_count': [128],
        'model__eval_metric': ['AUC'],
        'model__early_stopping_rounds': [50]
    }
]
```

```python
# RandomizedSearchCV для классификации
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=3,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs= -1,
    n_iter=50,
    return_train_score=True
)
```

```python
print("ПОДБОР ГИПЕРПАРАМЕТРОВ С КРОСС-ВАЛИДАЦИЕЙ")
randomized_search.fit(X_train, y_train)
```

```python
best_model = randomized_search.best_estimator_
best_cv_rmse = -randomized_search.best_score_
best_model_name = type(best_model.named_steps['model']).__name__
```

```python
print(f"\nРЕЗУЛЬТАТЫ АНАЛИЗА:")
print(f" Лучшая модель: {best_model_name}")
print(f" Лучший ROC-AUC score: {randomized_search.best_score_}")
print(f" Параметры лучшей модели: {randomized_search.best_params_}")
```

```python
best_model.fit(X_train, y_train)
```

```python
# Получим предсказания
y_pred_proba = best_model.predict_proba(X_test)[:, 1]
y_pred = best_model.predict(X_test)
```

```python
# Оценим качество на тестовой выборке 
test_roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"Тестовый ROC-AUC: {test_roc_auc:.4f}")
```
